```python
import numpy as np
import pandas as pd
from os import path

import matplotlib.pyplot as plt
import seaborn as sb
```

```python
dirname = '/Users/xies/Library/CloudStorage/OneDrive-Stanford/Skin/Mesa et al/Lineage models/Annotability results'

consistency = pd.read_csv(path.join(dirname,'feature_consistency_across_delays.csv'))
merged = pd.read_csv(path.join(dirname,'merged_cross_delay_summary.csv'))
correlates = pd.read_csv(path.join(dirname,'top_correlated_features_ALL_DELAYS.csv'))
```

```python
correlates[correlates['Feature'] == 'Mean curvature 2um']
```

```python
df = correlates[(correlates['Metric'] == 'Confidence') &
    ((correlates['Feature'] == 'Mean curvature 5um') |
     (correlates['Feature'] == 'Mean curvature 2um') |
     (correlates['Feature'] == 'Basal area') |
    (correlates['Feature'] == 'Mean curvature - cell coords') | 
    (correlates['Feature'] == 'Nuclear height'))]

g = sb.FacetGrid(df,row='Fate',col='Category',legend_out=True)
g.map_dataframe(sb.lineplot,x='Delay_hours',y='Correlation',hue='Feature',
               hue_order=['Mean curvature 5um','Mean curvature 2um','Basal area','Mean curvature - cell coords','Nuclear height'])
g.add_legend()
```
